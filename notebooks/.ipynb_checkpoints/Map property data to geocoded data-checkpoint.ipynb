```python
import numpy as np
import pandas as pd
import os
import glob
```

```python
df_property = load_csv = pd.read_csv('../combined_data/sydney_sales_combined.csv')
df_geocoded = load_csv = pd.read_csv('../geocoded_data/geocoded_address_list_master.csv')
```

```python
# function returns address separated into unit and street address

def break_address(row):
    a = row.ADDRESS.split('/')
    if len(a)==1:
        return pd.Series({'unit': None, 'street_address': a[0]})    
    return pd.Series({'unit': a[0], 'street_address': a[1]})
```

```python
df_addresses = df_property.apply(break_address, axis=1)
```

```python
df_property_two = df_property.merge(df_addresses, left_index=True, right_index=True, how='left')

df_property_two.head(1)
```

```python
df_geocoded.head(1)
```

```python
df_combined = df_property_two.merge(df_geocoded, on='street_address', how='left')

df_combined[40000:41500].head(20)
```

```python
df_combined.to_csv('../combined_data/sydney_sales_combined_geocoded.csv')
```
